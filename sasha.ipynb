```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
from bs4 import BeautifulSoup
```

```python
records = []
with open('data/bigfoot_first100records.json') as f:
    for i in f:
        records.append(json.loads(i))
```

```python
test = BeautifulSoup(records[0]['html'], 'html.parser')
test.find_all('p')[:-4]
```

```python
d = {'YEAR':[], 'SEASON':[], 'MONTH':[], 'STATE':[], 'COUNTY':[], 'LOCATION DETAILS':[], 'NEAREST TOWN':[], 'NEAREST ROAD':[], 'OBSERVED':[], 'ALSO NOTICED':[],
    'OTHER WITNESSES':[], 'OTHER STORIES':[], 'TIME AND CONDITIONS':[], 'ENVIRONMENT':[]}
for i in range(len(records[:3])):
    soup = BeautifulSoup(records[i]['html'], 'html.parser')
    for k in d.keys():
        for field in soup.find_all('p')[:-4]:
            obs = field.text.split(':', max_split=1)
            if k in obs:
                d[k].append(obs[1])
            d[k].append('None')
        
        
```

```python
for k, v in d.items():
    print(k, v)
```
