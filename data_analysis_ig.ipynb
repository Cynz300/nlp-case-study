```python
from tqdm import tqdm
bigfoot = []
with open('data/bigfoot_data.json') as f:
    for i in tqdm(f):
        bigfoot.append(json.loads(i))
        
big = {}

for record in tqdm(bigfoot):
    soup = BeautifulSoup(record['html'], 'html.parser')
    for para in soup.find_all('p'):
        data = para.text.split(':')
        if len(data) < 2 or (not data[0].isupper()):
            continue
        else:
            try:
                if big.get(data[0]) is None:
                    big[data[0]] = data[1:]
                else:
                    big[data[0]].append(data[1:])
            except IndexError:
                print(para)
                continue
                
big_keys = big.keys()

bigfoot_df = pd.DataFrame(None, index = range(len(bigfoot)), columns = big_keys)

for idx, record in tqdm(enumerate(bigfoot)):
    soup = BeautifulSoup(record['html'], 'html.parser')
    for para in soup.find_all('p'):
        data = para.text.split(':', maxsplit=1)
        if len(data) < 2 or (not data[0].isupper()):
            continue
        else:
            try:
                bigfoot_df.loc[idx, data[0]] = data[1]
            except IndexError:
                print(para)
                continue
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import nltk
from nltk.corpus import stopwords

tfidf = TfidfVectorizer(stop_words='english')
observations = tfidf.fit_transform(bigfoot_df['OBSERVED']).toarray()
```

```python
new_series = bigfoot_df['STATE'].copy()

print(type(new_series))
for idx, value in enumerate(bigfoot_df['STATE']):
    if type(value) == float and idx > 0:
        new_series[idx] = new_series[idx - 1]
        
bigfoot_df['STATE'] = new_series
```

```python
import numpy as np
keys = []
values = []
for key, value in zip(bigfoot_df['STATE'].value_counts().index, bigfoot_df['STATE'].value_counts()):
    keys.append(key)
    values.append(value)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize = (10,10))
plt.bar(keys, values)
plt.ylabel('Instances')
plt.xlabel('States')
plt.xticks(rotation = 90)
plt.title('Bigfoot sightings by State')
plt.savefig('bigfoot_state.png');
```
