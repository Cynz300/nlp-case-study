```python
from tqdm import tqdm
bigfoot = []
with open('data/bigfoot_data.json') as f:
    for i in tqdm(f):
        bigfoot.append(json.loads(i))
        
big = {}

for record in tqdm(bigfoot):
    soup = BeautifulSoup(record['html'], 'html.parser')
    for para in soup.find_all('p'):
        data = para.text.split(':')
        if len(data) < 2 or (not data[0].isupper()):
            continue
        else:
            try:
                if big.get(data[0]) is None:
                    big[data[0]] = data[1:]
                else:
                    big[data[0]].append(data[1:])
            except IndexError:
                print(para)
                continue
                
big_keys = big.keys()

bigfoot_df = pd.DataFrame(None, index = range(len(bigfoot)), columns = big_keys)

for idx, record in tqdm(enumerate(bigfoot)):
    soup = BeautifulSoup(record['html'], 'html.parser')
    for para in soup.find_all('p'):
        data = para.text.split(':', maxsplit=1)
        if len(data) < 2 or (not data[0].isupper()):
            continue
        else:
            try:
                bigfoot_df.loc[idx, data[0]] = data[1]
            except IndexError:
                print(para)
                continue
```
